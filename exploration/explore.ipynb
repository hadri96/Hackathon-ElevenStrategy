```python
import pandas as pd
import matplotlib.pyplot as plt

# Supposons que votre DataFrame s'appelle 'df' et que vous l'avez déjà créé
# à partir de vos fichiers CSV ou XLSX en utilisant le code précédent.
# Si ce n'est pas le cas, vous devrez d'abord importer vos données.

# Exemple, si votre dataframe s'appelle dataframe_csv (adapter le nom si besoin):
df = attendance # ou dataframe_xlsx, ou le nom que vous avez donné à votre dataframe

# 1. Filtrer le DataFrame pour ne garder que "PortAventura World" dans FACILITY_NAME
portaventura_data = df[df['FACILITY_NAME'] == 'PortAventura World'].copy() # Utilisez .copy() pour éviter les warnings lors de modifications

# 2. Convertir la colonne 'USAGE_DATE' en type datetime
portaventura_data.loc[:, 'USAGE_DATE'] = pd.to_datetime(portaventura_data['USAGE_DATE']) # Use .loc to avoid SettingWithCopyWarning

# 3. Filtrer par date pour les périodes souhaitées
date_debut_periode1 = pd.to_datetime('2018-01-01')  # Date de début de la première période (adapter si besoin)
date_fin_periode1 = pd.to_datetime('2019-12-31')
date_debut_periode2 = pd.to_datetime('2022-01-01')
date_fin_periode2 = pd.to_datetime('2024-12-31') # ou une date future si vous voulez inclure toutes les données après début 2022

periode1_data = portaventura_data[(portaventura_data['USAGE_DATE'] >= date_debut_periode1) & (portaventura_data['USAGE_DATE'] <= date_fin_periode1)]
periode2_data = portaventura_data[(portaventura_data['USAGE_DATE'] >= date_debut_periode2) & (portaventura_data['USAGE_DATE'] <= date_fin_periode2)]

# 4. Concaténer les données des deux périodes
data_a_plotter = pd.concat([periode1_data, periode2_data]).sort_values(by='USAGE_DATE') # Trier de nouveau pour assurer l'ordre chronologique

# 5. Créer le plot en utilisant les données filtrées
plt.figure(figsize=(12, 6)) # Ajuster la taille du graphique si nécessaire
plt.plot(data_a_plotter['USAGE_DATE'], data_a_plotter['attendance'])

# Ajouter des titres et labels
plt.title('Attendance à PortAventura World (2018-2019 et à partir de 2022)')
plt.xlabel('Date')
plt.ylabel('Attendance')

# Formatter l'axe des dates (optionnel)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Afficher le plot
plt.show()
```
