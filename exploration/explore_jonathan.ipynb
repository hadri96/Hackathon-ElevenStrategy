```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def prepare_data(df, target_ride):
    # Filter for specific ride
    df = df[df['ENTITY_DESCRIPTION_SHORT'] == target_ride].copy()
    
    # Create time-based features
    df['hour'] = df['DEB_TIME'].dt.hour
    df['day_of_week'] = df['WORK_DATE'].dt.dayofweek
    df['month'] = df['WORK_DATE'].dt.month
    df['is_weekend'] = df['WORK_DATE'].dt.dayofweek.isin([5, 6]).astype(int)
    
    # Drop columns we won't use for prediction
    columns_to_drop = ['ENTITY_DESCRIPTION_SHORT', 'DEB_TIME', 'FIN_TIME', 
                      'WORK_DATE', 'USAGE_DATE']
    df = df.drop(columns=columns_to_drop)
    
    # Separate features and target
    X = df.drop('WAIT_TIME_MAX', axis=1)
    y = df['WAIT_TIME_MAX']
    
    return X, y

def tune_hyperparameters(X, y):
    # Define parameter grid
    param_grid = {
        'n_estimators': [100, 200], #'n_estimators': [100, 200, 300],
        'max_depth': [3, 6], #[3, 6, 9, 12]
        'learning_rate': [ 0.1, 0,3],# [0.01, 0.1, 0.3]
        'min_child_weight': [3],#[1, 3, 5]
        'subsample': [0.9],#[0.8, 0.9, 1.0]
        'colsample_bytree': [0.9],#[0.8, 0.9, 1.0]
        'gamma': [0.1]#[0, 0.1, 0.2]
    }
    
    # Initialize model
    model = XGBRegressor(
        objective='reg:squarederror',
        random_state=42
    )
    
    # Setup GridSearchCV with 5-fold cross-validation
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grid,
        cv=2,
        scoring='neg_root_mean_squared_error',
        n_jobs=-1,
        verbose=1
    )
    
    # Fit GridSearchCV
    grid_search.fit(X, y)
    
    print("\nBest parameters found:")
    print(grid_search.best_params_)
    print(f"\nBest RMSE: {-grid_search.best_score_:.2f}")
    
    return grid_search.best_estimator_, grid_search.best_params_

def evaluate_with_cv(X, y, model, n_splits=5):
    # Initialize metrics storage
    cv_scores = {
        'rmse': [],
        'mae': [],
        'r2': []
    }
    
    # Initialize K-Fold
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    
    # Perform k-fold cross-validation
    for fold, (train_idx, val_idx) in enumerate(kf.split(X), 1):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
        
        # Train model
        model.fit(X_train, y_train)
        
        # Make predictions
        y_pred = model.predict(X_val)
        
        # Calculate metrics
        cv_scores['rmse'].append(np.sqrt(mean_squared_error(y_val, y_pred)))
        cv_scores['mae'].append(mean_absolute_error(y_val, y_pred))
        cv_scores['r2'].append(r2_score(y_val, y_pred))
        
        print(f"\nFold {fold} Results:")
        print(f"RMSE: {cv_scores['rmse'][-1]:.2f}")
        print(f"MAE: {cv_scores['mae'][-1]:.2f}")
        print(f"RÂ²: {cv_scores['r2'][-1]:.3f}")
    
    # Calculate average metrics
    avg_scores = {
        'avg_rmse': np.mean(cv_scores['rmse']),
        'std_rmse': np.std(cv_scores['rmse']),
        'avg_mae': np.mean(cv_scores['mae']),
        'std_mae': np.std(cv_scores['mae']),
        'avg_r2': np.mean(cv_scores['r2']),
        'std_r2': np.std(cv_scores['r2'])
    }
    
    return cv_scores, avg_scores

def train_and_evaluate_final_model(X, y, best_params, test_size=0.2):
    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
    
    # Initialize model with best parameters
    model = XGBRegressor(**best_params, random_state=42)
    
    # Train model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Calculate metrics
    train_metrics = {
        'rmse': np.sqrt(mean_squared_error(y_train, y_train_pred)),
        'mae': mean_absolute_error(y_train, y_train_pred),
        'r2': r2_score(y_train, y_train_pred)
    }
    
    test_metrics = {
        'rmse': np.sqrt(mean_squared_error(y_test, y_test_pred)),
        'mae': mean_absolute_error(y_test, y_test_pred),
        'r2': r2_score(y_test, y_test_pred)
    }
    
    # Get feature importance
    feature_importance = pd.DataFrame({
        'feature': X_train.columns,
        'importance': model.feature_importances_
    }).sort_values('importance', ascending=False)
    
    return {
        'model': model,
        'X_train': X_train,
        'X_test': X_test,
        'y_train': y_train,
        'y_test': y_test,
        'y_train_pred': y_train_pred,
        'y_test_pred': y_test_pred,
        'train_metrics': train_metrics,
        'test_metrics': test_metrics,
        'feature_importance': feature_importance
    }

def plot_results(results, title_prefix=""):
    # Create a figure with subplots
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
    
    # 1. Actual vs Predicted (Training Set)
    ax1.scatter(results['y_train'], results['y_train_pred'], alpha=0.5)
    ax1.plot([0, max(results['y_train'])], [0, max(results['y_train'])], 'r--')
    ax1.set_xlabel('Actual Wait Time')
    ax1.set_ylabel('Predicted Wait Time')
    ax1.set_title(f'{title_prefix}Training Set: Actual vs Predicted')
    
    # 2. Actual vs Predicted (Test Set)
    ax2.scatter(results['y_test'], results['y_test_pred'], alpha=0.5)
    ax2.plot([0, max(results['y_test'])], [0, max(results['y_test'])], 'r--')
    ax2.set_xlabel('Actual Wait Time')
    ax2.set_ylabel('Predicted Wait Time')
    ax2.set_title(f'{title_prefix}Test Set: Actual vs Predicted')
    
    # 3. Residuals Plot (Test Set)
    residuals = results['y_test'] - results['y_test_pred']
    ax3.scatter(results['y_test_pred'], residuals, alpha=0.5)
    ax3.axhline(y=0, color='r', linestyle='--')
    ax3.set_xlabel('Predicted Wait Time')
    ax3.set_ylabel('Residuals')
    ax3.set_title(f'{title_prefix}Residuals vs Predicted Values')
    
    # 4. Feature Importance Plot
    top_features = results['feature_importance'].head(10)
    sns.barplot(x='importance', y='feature', data=top_features, ax=ax4)
    ax4.set_title(f'{title_prefix}Top 10 Feature Importance')
    
    plt.tight_layout()
    plt.show()

def main():
    # Select a specific ride
    target_ride = "Roller Coaster"
    
    # Prepare the data
    X, y = prepare_data(merged_attendance_notnan, target_ride)
    
    print("Starting hyperparameter tuning...")
    best_model, best_params = tune_hyperparameters(X, y)
    
    print("\nPerforming cross-validation with best model...")
    cv_scores, avg_scores = evaluate_with_cv(X, y, best_model)
    
    print("\nCross-validation Results:")
    print(f"Average RMSE: {avg_scores['avg_rmse']:.2f} Â± {avg_scores['std_rmse']:.2f}")
    print(f"Average MAE: {avg_scores['avg_mae']:.2f} Â± {avg_scores['std_mae']:.2f}")
    print(f"Average RÂ²: {avg_scores['avg_r2']:.3f} Â± {avg_scores['std_r2']:.3f}")
    
    print("\nTraining final model with best parameters...")
    final_results = train_and_evaluate_final_model(X, y, best_params)
    
    print("\nFinal Model Results:")
    print("\nTraining Metrics:")
    print(f"RMSE: {final_results['train_metrics']['rmse']:.2f}")
    print(f"MAE: {final_results['train_metrics']['mae']:.2f}")
    print(f"RÂ²: {final_results['train_metrics']['r2']:.3f}")
    
    print("\nTest Metrics:")
    print(f"RMSE: {final_results['test_metrics']['rmse']:.2f}")
    print(f"MAE: {final_results['test_metrics']['mae']:.2f}")
    print(f"RÂ²: {final_results['test_metrics']['r2']:.3f}")
    
    # Plot results
    plot_results(final_results, f"{target_ride} - ")
    
    return {
        'best_model': best_model,
        'best_params': best_params,
        'cv_scores': cv_scores,
        'avg_scores': avg_scores,
        'final_results': final_results
    }

if __name__ == "__main__":
    results = main()
```
